```python
from requests_html import HTMLSession
```

```python
trade = "Limestone Bancorp, Inc. - Common Stock [LMST]"
```

```python
def cleanNewsURLQuery(trade):
    return (
        'https://news.google.com/search?q={}&hl=en-US&gl=US&ceid=US%3Aen'.format(
            trade.replace(' ', '%20').replace(',', '').replace('[','%5B').replace(']','%5D')
            )
        )

def fetchSession(url):
    session = HTMLSession()
    r = session.get(url)
    return r
```

```python
url = cleanNewsURLQuery(trade)
print(url)
```

```python
r = fetchSession(url)
```

```python
page = r.html.find('main')
```

```python
type(page)
```

```python
elements = page[0].find('div')
```

```python
for e in elements:
    print(e.text)
```
